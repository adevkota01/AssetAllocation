```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.style as style
style.use('fivethirtyeight')
%matplotlib inline
```

```python
df = pd.read_excel('data/RF_GS_UP_HistoricalResults.xlsx')
plt.figure(figsize=(30,20))
plt.tight_layout()
snsPlot = sns.regplot(x=df["Prob"], y=df["Target_FR12Mts"], fit_reg=False, scatter_kws={"color":"blue","alpha":0.9,"s":200})
plt.xlabel('Historical Probability', fontsize=35)
plt.ylabel('12Month Forward Returns', fontsize=35)
plt.axhline(0, color="red", linestyle="--")
plt.axhline(0.08, color="green", linestyle="--")


plt.axvline(0.5, color="red", linestyle="--")

plt.axvline(0.02, color="orange", linestyle="--")

plt.title('Random Forest Grid Search / Forward 12M Return > 8% \n\n Mar 2020 Probability 0.02', fontsize=40)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)

# plt.show()

fig = snsPlot.get_figure()
fig.savefig('../static/img/portfolio/RFGS1.jpg')
```
